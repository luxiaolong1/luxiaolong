```python
data = np.random.randint(150,size=(3,3))
index = ['张三','李四','王老五']
columns = ['语文','数学','英语']
ddd = DataFrame(data=data,index = index,columns=columns)
ddd
```

```python
data = np.random.randint(150,size=(2,3))
index = ['张十三','赵老六']
columns = ['语文','数学','英语']
ddd2 = DataFrame(data=data,index = index,columns=columns)
ddd2
```

```python
ddds = pd.concat((ddd,ddd2))
ddds
```

```python
dds = ddds.rename(index={'张十三':'张三'})
dds
```
