```python
aapl = pd.read_csv(r'../data/AAPL.csv')
aapl.head()
```

```python
s_abbrevs = pd.read_csv(r'../data/state-abbrevs.csv')
s_abbrevs.head()
```

```python
s_areas = pd.read_csv(r'../data/state-areas.csv')
s_areas.head()
```

```python
s_pop = pd.read_csv(r'../data/state-population.csv')
s_pop.head()
```

## 合并pop与abbrevs两个DataFrame,分别依据state/region列和abbreviation列来合并.为了保留所有信息,使用外合并.

```python
temp1 = pd.merge(s_pop, s_abbrevs, how='outer',left_on='state/region', right_on='abbreviation')
temp1
```

## 去除abbreviation的那一列(axis=1)

```python
temp1.drop(['abbreviation'], axis=1,inplace=True)
temp1
```

## 查看存在缺失数据的列.使用.isnull().any(),只有某一列存在一个缺失数据,就会显示True.

```python
temp1.isnull().any(axis=0)
```

## 查看缺失数据,根据数据是否缺失情况显示数据

```python
# population缺失
lost1 = temp1['population'].isnull()
```

```python
temp1.loc[lost1]
```

```python
# state缺失
lost2 = temp1['state'].isnull()
```

```python
temp2 = temp1.loc[lost2]
temp2
```

## 找到有哪些state/region的state的值为NaN，使用unique()查看非重复值

```python
temp2['state/region'].unique()
```

## 给找到的这些state/region的state项补上正确的值,从而去除掉state这一列的所有NaN

```python
# 对结果有影响, state的类型是object
temp1.loc[temp1['state/region']=='PR', 'state'] = 'Puerto Rico'
temp1.loc[temp1['state/region']=='USA', 'state'] = 'United States'
temp1
```

```python
temp1.dtypes
```

```python
# 只要temp1没有空数据就好
temp1['state'].isnull().any()
```

## 合并各州面积数据areas,使用左合并.

```python
abb_pop_areas = pd.merge(temp1, s_areas, how='left')
abb_pop_areas
```

```python
为什么使用外合并?
左表含有的有用信息较多(year, ages等),而右表是固定的.
使用左合并,能得到更多的信息.
如:各个年份的各州的人口密度.
在数据分析中,一开始我们可能不完全了解哪些数据是有用的,因此应该尽量分析更多的数据.
```

## 继续寻找存在缺失数据的列

```python
abb_pop_areas.isnull().any()
```

## 我们会发现area(sq.mi)这一列有缺失数据，为了找出是哪一行，我们需要找出是哪个state没有数据

```python
list = abb_pop_areas.loc[abb_pop_areas['area (sq. mi)'].isnull()].index
list
```

## 去除含有缺失数据的行

```python
abb_pop_areas.drop(index=list, axis=0, inplace=True)
```

```python
abb_pop_areas
```

```python
abb_pop_areas.dtypes
```

## 查看数据是否缺失

```python
abb_pop_areas.isnull().any()
```

## 找出2010年的全民人口数据,df.query(查询语句)

```python
pop_state_total_2010 = abb_pop_areas.query("year==2010 & ages=='total'")
```

## 对查询结果进行处理，以state列作为新的行索引:set_index

```python
final_data = pop_state_total_2010.set_index(pop_state_total_2010['state'])
```

```python
final_data
```

 ## 计算人口密度.注意是Series/Series,其结果还是一个Series.

```python
data =final_data['population']  / final_data['area (sq. mi)'] 
type(data)
```

```python
population_density = pd.Series(data=data, index=final_data['state'])
```

```python
population_density
```

```python
final_data['population_density'] = population_density
```

```python
final_data
```

## 排序，并找出人口密度最高的五个州sort_values()

```python
final_data.sort_values('population_density')
```

```python
final_data.sort_values('population_density').tail()
```
